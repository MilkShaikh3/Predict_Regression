```python
import numpy as np
import pandas as pd 

#Loading scikit's random foresst Classifier library
from sklearn.ensemble import RandomForestClassifier
#setting random seed
np.random.seed(0)
```

```python
#read in data 
urltrain= 'https://raw.githubusercontent.com/MilkShaikh3/Predict_Regression/master/Train.csv'
urltest = 'https://raw.githubusercontent.com/MilkShaikh3/Predict_Regression/master/Test.csv'
traindata = pd.read_csv(urltrain)
testdata = pd.read_csv(urltest)
```

```python
# One-hot encode the data using pandas get_dummies
traindata = pd.get_dummies(traindata)
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(traindata['Time from Pickup to Arrival'])
# Remove the labels from the features
# axis 1 refers to the columns
traindata = traindata.drop('Time from Pickup to Arrival', axis = 1)

```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(traindata, labels, test_size = 0.25, random_state = 42)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 10, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels)

#output :MemoryError: Unable to allocate 12.3 GiB for an array with shape (103942, 15900) and data type float64
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
```
